```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Alice", "David"],
    "Age": [25, 30, 25, 40],
    "City": ["NY", "LA", "NY", "Chicago"],
    "time": ["Lunch", "Dinner", "Lunch", "Dinner"],
    "total_bill": [20.5, None, 20.5, 50.0]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

df_cleaned = df.drop_duplicates()

print("\nModified DataFrame (no duplicates)")
print(df_cleaned)
```

```python
df.pivot_table(index=['day'], columns=['time'], values='total_bill', aggfunc='mean', fill_value=0, margins=True)
```
