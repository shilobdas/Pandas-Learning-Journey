```python
import matplotlib.pyplot as plt
```

```python
#data
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y)
plt.show()
```

# 2 Customize Chart

```python
plt.figure(figsize=(8,5),dpi=100)#figure size
plt.plot(x,y,'o--b',linewidth=2,markersize=12)
plt.title("Line chart Customization")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

#3 Advance Line chart

```python
#data
x=[1,2,3,4,5]
y1=[10,20,25,35,45]
y2=[20,30,35,45,55]
plt.plot(x, y1, label='sale 2024')#ploting y1 data
plt.plot(x, y2, label='sale 2025')#ploting y2 data
plt.title("YoY Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()#TO Show Legend
plt.show()
```

#4 bar Chart

```python
#Data
x=['A','B','C','D','E']
y=[10,20,25,50,45]
plt.bar(x,y)
plt.title("Bar Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

#5 Histogram

```python
#Used for distribution analysis
#data
import random
data =[random.randint(1,10) for _ in range(100)]
plt.hist(data,bins=20)
plt.title("Histogram Example")
plt.show()
```

#6 Pie Chart

```python
#Used to show part-to-whole reletionship/data
#Data
categories=['A','B','C','D','E']
sales=[10,20,55,35,45]
plt.pie(sales,labels=categories,autopct='%1.1f%%',startangle=50)
plt.title("Pie Chart")
plt.legend()
plt.show("Pie chart")
```

#7 Scatter plot

```python
#used to find reletionship btwn variabls
#data
y1=[10,20,25,35,45]
y2=[20,30,35,45,55]
plt.scatter(y1,y2)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

#8 Subplots

```python
import matplotlib.pyplot as plt
#used to show multiple charts in one figure
#data1-barchart
categories=['Mon','Tue','Wed','Thu','Fri']
sales=[10,20,55,35,45]

#deta2-scater plot
y1=[10,20,25,35,45]
y2=[20,30,35,45,55]

plt.figure(figsize=(10,6))
#First plot-barchart
plt.subplot(1,2,2)# row column ,position
plt.bar(categories,sales)
plt.title("Weekly Sales")
plt.xlabel("week Days")
plt.ylabel("Sales")

#second plot -sccater chart
plt.subplot(1,2,1)# row column ,position
plt.scatter(y1,y2)
plt.title("user example")
plt.xlabel("User1 ")
plt.ylabel("User2")
plt.show()
```
