```python
import matplotlib.pyplot as plt
```

```python
#data
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y)
plt.show()
```

# 2 Customize Chart

```python
plt.figure(figsize=(8,5),dpi=100)#figure size
plt.plot(x,y,'o--b',linewidth=2,markersize=12)
plt.title("Line chart Customization")
plt.show()
```
